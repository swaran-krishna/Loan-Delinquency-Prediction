```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score
import xgboost as xgb
import matplotlib.pyplot as plt
from xgboost import XGBClassifier

# 1. Generate Synthetic Data (replace with real data)
np.random.seed(42)
n_samples = 1000

data = {
    'Income': np.random.normal(70000, 25000, n_samples).clip(20000, 150000),
    'Credit_Utilization': np.random.beta(2, 5, n_samples) * 1.5,  # Some values >1.0
    'Missed_Payments_12M': np.random.poisson(0.7, n_samples).clip(0, 5),
    'Age': np.random.randint(18, 70, n_samples),
    'Credit_Score': np.random.normal(650, 100, n_samples).clip(300, 850),
    'Debt_to_Income_Ratio': np.random.beta(2, 3, n_samples) * 0.6
}

# Create target variable (delinquency)
data['Delinquent'] = (
    (data['Credit_Utilization'] > 0.8) | 
    (data['Missed_Payments_12M'] >= 2) |
    ((data['Income'] < 50000) & (data['Debt_to_Income_Ratio'] > 0.4))
).astype(int)

# Add some noise
data['Delinquent'] = data['Delinquent'] ^ (np.random.random(n_samples) < 0.05)

df = pd.DataFrame(data)

# 2. Data Preprocessing
# Handle outliers
df['Credit_Utilization'] = df['Credit_Utilization'].clip(0, 1.0)

# Feature engineering
df['Utilization_to_Income'] = df['Credit_Utilization'] / (df['Income']/1e5)
df['Payment_Stability'] = 1 / (df['Missed_Payments_12M'] + 1)

# Select features
features = [
    'Income', 
    'Credit_Utilization',
    'Missed_Payments_12M',
    'Credit_Score',
    'Debt_to_Income_Ratio',
    'Utilization_to_Income',
    'Payment_Stability'
]
X = df[features]
y = df['Delinquent']

# 3. Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)

# Corrected XGBoost implementation
model = xgb.XGBClassifier(
    objective='binary:logistic',
    scale_pos_weight=sum(y_train==0)/sum(y_train==1),
    max_depth=5,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.7,
    eval_metric='aucpr',
    early_stopping_rounds=10  # MOVED THIS HERE
)

model.fit(
    X_train, y_train,
    eval_set=[(X_test, y_test)],  # Validation data for early stopping
    verbose=True
)

# 5. Evaluation
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]

print("Classification Report:")
print(classification_report(y_test, y_pred))

print("\nAUC-ROC:", roc_auc_score(y_test, y_proba))

# 6. Feature Importance
plt.figure(figsize=(10, 6))
xgb.plot_importance(model, max_num_features=10)
plt.title('Feature Importance')
plt.show()

# 7. Prediction Example
new_customer = pd.DataFrame([{
    'Income': 45000,
    'Credit_Utilization': 0.95,
    'Missed_Payments_12M': 3,
    'Credit_Score': 580,
    'Debt_to_Income_Ratio': 0.5,
    'Utilization_to_Income': 0.95/(45000/1e5),
    'Payment_Stability': 1/(3+1)
}])

print("\nNew Customer Prediction:")
print("Probability of Delinquency:", model.predict_proba(new_customer)[0][1])
print("Risk Category:", "High" if model.predict(new_customer)[0] == 1 else "Low")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns

# =========================================
# 1. Load cleaned data
# =========================================
df = pd.read_csv("cleaned_delinquency_data.csv")

# =========================================
# 2. Create payment history features
# =========================================
payment_cols = ['Month_1', 'Month_2', 'Month_3', 'Month_4', 'Month_5', 'Month_6']

df['Total_Late'] = df[payment_cols].apply(lambda x: (x == 'Late').sum(), axis=1)
df['Total_Missed'] = df[payment_cols].apply(lambda x: (x == 'Missed').sum(), axis=1)

# =========================================
# 3. Select features and target
# =========================================
features = ['Income', 'Credit_Utilization', 'Total_Missed', 'Total_Late',
            'Debt_to_Income_Ratio', 'Credit_Score']
target = 'Delinquent_Account'

X = df[features]
y = df[target]

# =========================================
# 4. Train/Test Split
# =========================================
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# =========================================
# 5. Feature Scaling
# =========================================
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# =========================================
# 6. Train Logistic Regression
# =========================================
log_reg = LogisticRegression(class_weight='balanced', random_state=42)
log_reg.fit(X_train_scaled, y_train)

# =========================================
# 7. Predictions
# =========================================
y_pred = log_reg.predict(X_test_scaled)
y_pred_proba = log_reg.predict_proba(X_test_scaled)[:, 1]

# =========================================
# 8. Evaluation
# =========================================
print("Classification Report:")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:")
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

roc_auc = roc_auc_score(y_test, y_pred_proba)
print(f"ROC-AUC Score: {roc_auc:.3f}")

# =========================================
# 9. ROC Curve
# =========================================
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
plt.figure(figsize=(6,4))
plt.plot(fpr, tpr, label=f"Logistic Regression (AUC={roc_auc:.2f})")
plt.plot([0,1], [0,1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```
